# Data Visualization

- Data visualization libraries
    - (1) Matplotlib
        - Matplotlib is a Python 2D plotting library which produces publication quality figures in a variety of hardcopy formats and interactive environments across platforms.
    - (2) Seaborn : 
        - Seaborn is a Python data visualizationh library based on matplotlib. It provides high-level interface for drawing attracttive and informative statistical graphics.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Plot series with different magnitudes

```python
t = s.copy()
t1 = (pd.Series(np.random.randn(10).cumsum(), index = np.arange(0,100,10)))/100
```

```python
plt.figure(figsize=(12,8))
plt.plot(t, label='t')
plt.plot(t1, color = 'red', label='t1')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Example')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,8))
ax1 = plt.subplot()
ax1.plot(t, label='t')
ax2 = ax1.twinx()
ax2.plot(t1, color = 'red', label='t1')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Example')
plt.legend()
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12,8))
ax1 = plt.subplot()
ax1.plot(t, label='t')
ax2 = ax1.twinx()
ax2.plot(t1, color = 'red', label='t1')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Example')
plt.legend()
plt.show()
```

(2) Bar plot

```python
plt.figure(figsize=(12,8))
plt.bar(x=s.index, height=s, label='s')
plt.bar(x=s1.index, height=s1, color = 'red', label='s')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Example')
plt.legend()
plt.show()
```

(3) Scatterplot

```python
plt.figure(figsize=(12,8))
s=np.random.randn(1000)
s1=np.random.randn(1000)

plt.scatter(s,s1)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Example')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

ols = LinearRegression()
ls = ols.fit(s.reshape(1000,1),s1.reshape(1000,1))
s2 = ls.predict(s1.reshape(1000,1))
#s2.index = s1
```

```python
plt.figure(figsize=(12,8))
plt.scatter(s1,s, label='scatter')
plt.plot(pd.Series(s2.reshape(1000,),index=s1), color='r', label='Regression Line')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Example')
plt.legend()
plt.show()

# sns.lmplot()
```

## Seaborn

(1) Distplot

```python
titanic = sns.load_dataset('titanic', engine='python', )
titanic.head()
```

```python
sns.distplot(titanic.fare)
```

(2) Jointplot

```python
sns.jointplot(x=titanic.age, y=titanic.fare)
# kind = 'hex'
```

(3) Pairplot

```python
titanic['gender'] = 0
titanic.loc[titanic.sex=='male', 'gender'] = 1
```

```python
sns.pairplot(titanic[['survived', 'fare', 'gender', 'age']], hue='survived')
```
